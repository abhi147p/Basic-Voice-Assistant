```python
import pyttsx3
import speech_recognition as sr
from datetime import datetime
import wikipedia
import webbrowser
import os
from MovieRecom import movie_recommender
from BookRecom import book_recommender

engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

# print(voices)

def wishMe(name):
    hour = int(datetime.now().hour)
    if hour>=0 and hour<12:
        speak(f"Good Morning! {name}")
    elif hour>=12 and hour<18:
        speak(f"Good Afternoon! {name}")  
    else:
        speak(f"Good Evening! {name}")  
    # speak("I am Anand. Please tell me how may I assist you")      

def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.record(source,duration=5)
    try:
        print("Recognizing...")    
        query = r.recognize_google(audio,language='en-in')
        print(f"User said: {query}\n")
        #ok=1
    except Exception as e:
        speak("Say that again please...")
        return None
    return query

def Movie_Recommenders():
    
    speak('Search for any similar Movie?')
    print('Search for any similar Movie?')
    
    q = takeCommand().lower()
    if 'go back' in q:
        return
    if 'yes' in q or 'yeah' in q:
        speak('Tell Movie Name')
        print('Tell Movie Name')
        mov_name = takeCommand().lower()
        # ok=1
        recom_movies = movie_recommender(mov_name)
        print('Recommended Movies for:'+str(mov_name))
        speak('Recommended Movies for'+str(mov_name))
        print(recom_movies)
    elif 'no' in q or 'nope' in q:
        recom_movies = movie_recommender()
        print('Recommended movies are:')
        speak('Recommended movies are')
        # ok=1
        print(recom_movies)
#     if ok==0:
#         Movie_Recommenders(1)
    return

def Book_Recommenders():
    
    speak('Search for any similar Book?')
    print('Search for any similar Book?')
    
    q = takeCommand().lower()
    
    if 'go back' in q:
        return
    if 'yes' in q or 'yeah' in q:
        speak('Tell Book Name')
        print('Tell Book Name')
        bk_name = takeCommand().lower()
        book_recom = book_recommender(bk_name)
        print('Recommended Books for:'+str(bk_name))
        speak('Recommended Books for'+str(bk_name))
        for book in book_recom:
            print(book)
    elif 'no' in q or 'nope' in q:
        book_recom = book_recommender()
        print('Recommended Books are:')
        speak('Recommended books are')
        for book in book_recom:
            print(book)
#     if ok==0:
#         Book_Recommenders(1)
    return

#_name_="main"
if __name__ == "__main__":
    speak("Hey! I'm Lee. Please tell me how may I assist you")
    speak("Before starting, please tell your name")
    name = takeCommand().lower()
    wishMe(name)
   
    while True:
        query = takeCommand()
        if query is None:
            continue
        else:
            query = query.lower()
        # query = "open google"
        if 'wikipedia' in query:
            speak('Searching Wikipedia...')
            #query = query.replace("wikipedia", "")
            sub_query = takeCommand().lower()
            results = wikipedia.summary(sub_query, sentences=2)
            speak("According to Wikipedia")
            print(results)
            speak(results)
        elif 'youtube' in query:
            speak('Searching youtube ...')
            webbrowser.open("youtube.com")
        elif 'google' in query:
            speak('Searching google ...')
            webbrowser.open("google.com")
        elif 'firefox' in query:
            speak('Searching firefox ...')
            webbrowser.open("firefox.com")
        elif 'hi' in query:
            speak('Hey! , What\'s up?')
        elif 'stackoverflow' in query:
            speak('Searching stackoverflow ...')
            webbrowser.open("stackoverflow.com")
        elif ('music' in query) or ('songs' in query) or ('song' in query):
            music_dir = r'C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Songs'
            songs=os.listdir(music_dir)
            print(songs)
            os.startfile(os.path.join(music_dir,songs[0]))
        elif 'exit' in query or 'quit' in query:
            print('Thank you, see you again, have a good day')
            speak('Thank you, see you again, have a good day')
            break
        elif 'who are you' in query:
            speak('Im Lee, the Voice assistant!')
        elif 'time' in query:
            strTime = datetime.now().strftime("%H:%M:%S")
            speak(f"the time is {strTime}")
        elif 'date' in query or 'day' in query:
            time = datetime.now()
            week, day, month, yr = time.strftime('%A'), time.day, time.month, time.year
            mapped_month = {
                1: "January",
                2: "February",
                3: "March",
                4: "April",
                5: "May",
                6: "June",
                7: "July",
                8: "August",
                9: "September",
                10: "October",
                11: "November",
                12: "December"
            }
            speak(f"It's {week} {day} of {mapped_month[month]}, {yr}")
        elif 'powerpoint' in query:
            codePath = r"C:\ProgramData\Microsoft\Windows\Start Menu\Programs\PowerPoint"
            os.startfile(codePath)
        elif 'notepad' in query:
            codePath = r"C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Notepad"
            os.startfile(codePath)
        elif 'zoom' in query:
            speak('opening zoom')
            webbrowser.open("zoom.com")
        elif 'facebook' in query:
            speak('opening facebook')
            webbrowser.open("facebook.com")
        elif ('instagram' in query) or ('insta' in query):
            speak('opening Instagram')
            webbrowser.open("instagram.com")
        elif 'recommend' in query or 'suggest' in query:
            if 'book' in query:
                Book_Recommenders()
            elif 'movie' in query:
                Movie_Recommenders()
```
